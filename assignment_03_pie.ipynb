#   #     Assignment #    3 - Domains pie chart

This notebook will read in the data from a googledrive link and display the domains used via a pie chart.


````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# creating list with the specific columns to be used
column_name=['Email']

# reading in the specified column from CSV file
df = pd.read_csv("https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK", usecols=column_name)
````

````python
# extracting the email domain info from the email column using the ``.str.split`` method and adding the domain information to it's own column
df['email_domain'] = df['Email'].str.split('@').str[1]
````

````python
# counting the amount of times each domain appears
domain_counts = df['email_domain'].value_counts()

# coverting the results for domain_counts to a list of tuples of the domain and the number of times it appears in the CSV file
domain_list = list(domain_counts.items())
````

````python
# creating an empty list to use for the labels of the pie chart
mylabels = []

# creating custom labels including count of each domain to be used for the legend
for domain, count in domain_list: # looping through the 'domain_list' list taking the domain and count information stored for each tuple
    mylabels.append(f'{domain}: {count}') # adding the values to the ``mylabels`` list defined above in the desired format

# creating specific colours for the piechart
mycolors = ["#BF40BF", "#D3D3D3", "#8ACE00"]
````

````python
# creating the pie chart
plt.pie(domain_counts, labels=domain_counts.index, autopct='%1.1f%%', textprops={'fontweight': 'bold'}, shadow = True, colors=mycolors) # using the ```.index``` to take the domain names from the ``domain_counts`` series automatically

plt.legend(title = "No. of entries per domain:", loc='upper left', labels=mylabels, bbox_to_anchor=(-0.25, 0), ncols = 3) # adding the legend with 3 columns to the figure using the custom labels created
plt.title('Emails per domain type', y=1.05, fontsize=16, ha='center', fontweight='bold') # setting the title for the figure, it's size and positioning on the figure
plt.show()
````

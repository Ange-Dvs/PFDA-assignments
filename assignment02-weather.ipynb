This juptyer notebook will readh in the csv file weatherreadings1.csv and plot the temperature over time for the "dryBulbTemperature_Celsius" field.

```python
#importing required modules for the assignment
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
#reading in the file from a seperate folder
df = pd.read_csv("C:/Users/angel/OneDrive/College/pfda/PFDA-courseware/assignment/weatherreadings1.csv") 
```

```python
#cleaning up the reportStartDateTime column to strip out the excess charaters
df['reportStartDateTime'] = df['reportStartDateTime'].str[:19].str.replace('T', ' ') #setting the length of characters in the string to 19 and taking the character T out of it and replacing with a space
```

```python
df['reportStartDateTime'] = pd.to_datetime(df['reportStartDateTime']) #converting the string values in 'reportStartDateTime' column to a pandas datetime object to allow us to treat the value as date and time information and not just a string of characters but date and time data
```

```python
#setting the relevant columns in the csv file to variables to be reused throughtout the exercise
time = df['reportStartDateTime']
temp = df['dryBulbTemperature_Celsius']
```

```python
fig, ax = plt.subplots() #creating a figure and set of axes for a plot

fig.set_size_inches(10, 5) #setting the size of the figure to better show the data
ax.plot(time, temp, ".") #plotting the time and temp data and seeting the marker type for the plot

ax.set_title('Temperature (°C) over time', fontsize=16, y=1.05) #setting the title for the image
ax.set_xlabel('Date (dd/mm/yy) & Time', fontsize=14, fontstyle='italic', labelpad=10) #setting label for the x-axis and adding some space and styling parameters
ax.set_ylabel('Temperature (°C)', fontsize=14, fontstyle='italic', labelpad=10) #setting label for the y-axis and adding some space and styling parameters

ax.xaxis.set_major_locator(mdates.HourLocator(interval=2))  #setting the ticks to be every 2 hours to avoid overcrowding of the figure
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d/%m/%y %H:%M')) #formatting the ticks to display in the format of a date on the x-axis and setting the order of the information so it should appear dd/mm/yy hh:mm

min_temp = temp.min() #getting the min value for temp and assigning it to variable min_temp
max_temp = temp.max() #getting the max value for temp and assigning it to variable max_temp

plt.figtext(0.89, 0.82, f'Maximum temp: {max_temp:.2f}°C\nMinimum temp: {min_temp:.2f}°C', ha='right', fontsize=8) #adding text to the image which contains the information of the max and min temperature in the dataset

plt.xticks(rotation=90) #rotating ticks labels to ensure they don't overlap and make them easier to read 

ax.set_facecolor('lavender')  #setting the background colour of the plot
plt.grid(color='white', linestyle='--', linewidth=1)  #setting the grid lines to white, dotted lines for better contrast and visiblity
plt.show() #displaying the plot
```

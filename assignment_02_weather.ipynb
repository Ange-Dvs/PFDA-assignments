This juptyer notebook will read in the csv file 'weatherreadings1.csv' and plot the temperature over time for the 'dryBulbTemperature_Celsius' field.

```python
# importing required modules for the assignment
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# creating a list containing the headers that are relevant from the CSV file
relevant_column_names=['reportStartDateTime', 'dryBulbTemperature_Celsius']

# reading in the data, selecting the specific columns and defining the reportStartDateTime column as a datetime64 object
df = pd.read_csv('https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/refs/heads/main/assignment/weatherreadings1.csv', usecols=relevant_column_names, parse_dates=['reportStartDateTime']) 
```

```python
# setting the relevant columns to variables to be reused throughtout the exercise
time = df['reportStartDateTime']
temp = df['dryBulbTemperature_Celsius']
```

```python
fig, ax = plt.subplots() # creating a figure and set of axes for a plot

fig.set_size_inches(10, 5) # setting the size of the figure to better show the data
ax.plot(time, temp, '.') # plotting the time and temp data and seeting the marker type for the plot

# setting the title for the image & labels for the x-axis & y-axis with styling parameters
ax.set_title('Temperature (°C) over time', fontsize=16, y=1.05) 
ax.set_xlabel('Date & Time', fontsize=14, fontstyle='italic', labelpad=10) 
ax.set_ylabel('Temperature (°C)', fontsize=14, fontstyle='italic', labelpad=10)

# getting the min and max temperature
min_temp = temp.min() 
max_temp = temp.max() 

# setting the ticks to show every 2 hours to avoid overcrowding of the figure
ax.xaxis.set_major_locator(mdates.HourLocator(interval=2))  

# formatting the ticks so it should appear dd/mm/yy hh:mm
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d/%m/%y\n%H:%M')) 
plt.xticks(rotation=50) # rotating ticks labels


plt.figtext(0.89, 0.82, f'Maximum temp: {max_temp:.2f}°C\nMinimum temp: {min_temp:.2f}°C', ha='right', fontsize=8) # adding text to the image which contains the information of the max and min temperature in the dataset

ax.set_facecolor('lavender')  # setting the background colour of the plot
plt.grid(color='white', linestyle='--', linewidth=1)  # setting the grid lines to white, dotted lines for better contrast and visiblity
plt.show() # displaying the plot
```
